```python
SUN_list = './videolist/Sun/Sun_videolist_byWeek.csv'
BBC_list = './videolist/BBC/BBC_videolist_byDay.csv'
DM_list = './videolist/DailyMail/DailyMail_videolist_byDay.csv'
GD_list = './videolist/Guardian/Guardian_videolist_byDay.csv'
Independent_list = './videolist/Independent/Independent_videolist_byDay.csv'
```

```python
import csv
import time
import pandas as pd

def filter_by_data(data_list,media_name):
    if media_name == "SUN":
        video_path = SUN_list
    elif media_name == "BBC":
        video_path = BBC_list
    elif media_name == "DM":
        video_path = DM_list
    elif media_name == "Independent":
        video_path = Independent_list
    else:
        video_path = GD_list
        
    videolist = pd.read_csv(video_path)
    videolist['newtime'] = videolist.apply(lambda x: time.mktime(time.strptime(x.time.split('T')[0],"%Y-%m-%d")) if x.time != "" else 0,axis=1)
    
    
    final_list = pd.DataFrame(columns = videolist.columns.values)
    strptime_list = [time.mktime(time.strptime(i,"%Y-%m-%d")) for i in data_list]
    
    for i in range(0,len(strptime_list),2):
        final_list = final_list.append(videolist[(videolist['newtime']>= strptime_list[i])&(videolist['newtime']<= strptime_list[i+1])])
        
    return final_list[['channelTitle','videoId','vidoeTitle','time','link']]

# data_list : pairs of data, media_name : SUN, BBC, DM, GD, Independent
final_list = filter_by_data(["2020-03-11","2020-03-20"],"Independent")
```
